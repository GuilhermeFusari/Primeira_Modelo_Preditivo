### mudar os valores de 0, para valores de 1-7 baseado nos dias da semana

```python
def Transforma(lista):
    x = 0
    while x <= 364:
        if lista[x] == 1:
            if lista[x+1] == 0:
                lista[x+1] = 2
            else:
                pass
        else:
            pass
        x+=1  
    x = 0
    while x <= 364:
        if lista[x] == 2:
            if lista[x+1] == 0:
                lista[x+1] = 3
            else:
                pass
        else:
            pass
        x+=1 
    x = 0
    while x <= 364:
        if lista[x] == 3:
            if lista[x+1] == 0:
                lista[x+1] = 4
            else:
                pass
        else:
            pass
        x+=1 
    x = 0
    while x <= 364:
        if lista[x] == 4:
            if lista[x+1] == 0:
                lista[x+1] = 5
            else:
                pass
        else:
            pass
        x+=1 
    x = 0
    while x <= 364:
        if lista[x] == 1:
            if lista[x+1] == 1:
                lista[x] = 7
            else:
                pass
        else:
            pass
        x+=1 
    x = 0
    while x <= 364:
        if lista[x] == 7:
            if lista[x-1] == 0:
                lista[x-1] = 6
            else:
                pass
        else:
            pass
        x+=1
    return 
```

```python
lista[0] , lista[1] = 5, 6
data['Final de Semana'] = pd.DataFrame(lista)
data['Final de Semana']
```
